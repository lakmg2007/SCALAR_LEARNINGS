###3.   Marginal Probability:
Probability of a single event happening (e.g., P(A) or P(B)).

###4.   Joint Probability:
Probability of two events happening together (e.g., P(A ‚à© B)).

###5.   Conditional Probability:
Conditional probability calculates the likelihood of an event given that another event has already occurred.

---


###6.   Bayes' Theorem :
Bayes' Theorem is a cornerstone of probability theory that helps us revise probabilities based on new evidence. It enables us to calculate conditional probabilities when the reverse probability and other relevant information are known.
###Formula for Bayes' Theorem
The formula for Bayes' Theorem is:
P(A‚à£B)=P(B‚à£A)‚ãÖP(A)P(B)

Where:

P(A‚à£B) : Probability of  A  occurring given that  B  has occurred (posterior).
P(B‚à£A) : Probability of  B  occurring given that  A  has occurred (likelihood).
P(A) : Probability of  A  occurring (prior).
P(B) : Probability of  B  occurring (evidence).

#Question 3: It is known that 70% people like plain tea, 60% people like ginger tea, and 30% like both.What percentage of the people like plain tea, but do not like ginger tea?

```python
#The probability formula used here is based on set subtraction from set theory
#Formula Used:P(A¬†only)=P(A)‚àíP(A‚à©B) - subtraction rule
#Where:
#A: People who like plain tea
#B: People who like ginger tea
#A‚à©B: People who like both plain and ginger tea
#A¬†only: People who like only plain tea (i.e., plain tea but not ginger tea)
P(Plain only)=P(Plain) - P(Plain ‚à© inger)
= 70%-30%
=40%
```

#Question 6: A rare disease affects only 1 out of 1000 people. Upon conducting a diagnostic test to detect it, it is found that 90% of the affected people get a "positive" test result and that 5% of the non-affected people also get "positive" test result. Given that a person tested positive, what is the probability that he is actually affected?

```python
#Positive and affected = 90%
#Positive and non-affected = 5%
#Bay's theorm
P(diseased‚à£Positive)= P(Positive‚à£diseased)‚ãÖP(diseased) /P(Positive)

#Now lets calculate P(Positive)
Probability a person has the disease =1/1000 = 0.001
Probability of a person not diseased =1- 0.001=0.999
P(Positive|diseased)= 0.9
P(Positive|not-diseased) = 0.05

P(diseased‚à£Positive)= P(Positive‚à£diseased)‚ãÖP(diseased) /P(Positive)
P(Positive) = P(Positive‚à£Disease)‚ãÖP(Disease)+P(Positive‚à£No¬†Disease)‚ãÖP(No¬†Disease)
            = 0.9 * 0.001 + 0.05 * 0.999
            = 0.0009+0.04995=0.05085
#Now applying Bay's theorm
P(diseased‚à£Positive)=0.9 * 0.001 /0.05085 0.0009/0.05085 = 0.0177 = 17.7 %
```

#Question 7 : It is known that 20% of Netflix users are premium subscribers.Also, 80% of premium subscribers watch at least one thriller movie per week, and 50% of basic subscribers watch at least one thriller movie per week.What is the probability that a Netflix subscriber who watches at least one thriller movie per week is a premium subscriber?

```python
P(premium) = 20% = 0.20
P(Basic) = 80% = 0.80
P(Thriller|Premium) = 80% = 0.80
P(Thriller|Basic) = 50% = 0.5
#Need to find P(Thriller)
P(Thriller) = P(Thriller|Premium).P(Premium) + P(Thriller|Basic).P(Basic)
            = (0.8 * 0.2) + (0.5 * 0.8)
            = 0.16 + 0.40
            = 0.56
#Bay's theorm
P(Premium|Thriller) = P(Thriller|Premium).P(Premium)/P(Thriller)
                    = 0.8 * 0.2 / 0.56
                    = 0.16/0.56
                    = 0.2857
                    = 28.57%





```
