# **Data Types & Functional Programming:**
loc,iloc, lists, arrays, sets, tuples, dictionaries, strings, List comprehension, map, filter, lambda, reduce


```python
import pandas as pd
df = pd.DataFrame({'Name':['lakshmi','Dhatrhi'], 'Age':[47,16]})

#loc : label-based
print(df.loc[0,'Name'])
#iloc: index-based
print(df.iloc[1,0])
```

```python
#lists
my_lists = [1,2,3]
print(my_lists[2])
```

```python
#Arrays
import numpy as np
arr = np.array([1,2,3])
print(arr[1])
```

```python
#sets
my_set = {1,2,3,4}
print(my_set)
```

```python
#tuples
my_tuples = (1,2,3)
print(my_tuples[1])
```

```python
#Dictionaries
my_dict = {'name':'Alice', 'age':25}
print(my_dict['name'])
```

```python
#Strings
my_string = "Hello"
print(my_string[1])
```

```python
#List Comprehension
squares = [x**2 for x in range(5)]
print(squares)
```

```python
#map
nums = [1,2,3]
squared = list(map(lambda x:x**2, nums))
print(squared)
```

```python
#filter
nums = [1,2,3,4]
even = list(filter(lambda x: x%2 == 0, nums))
print(even)
```

```python
#lambda
add = lambda x,y:x+y
print(add(2,3))
```

```python
#reduce
from functools import reduce
nums = [1,2,3,4]
product = reduce(lambda x,y:x*y,nums)
print(product)
```

# **Numpy functions:**
Apply, Stacking, Splitting, Broadcasting, Vectorisation, Slicing, Indexing, Masking, Sorting.


```python
#apply (via np.vectorize or np.apply_along_axis)
import numpy as np

arr = np.array([1, 2, 3, 4])
square = np.vectorize(lambda x:x**2)
print(square(arr))
```

```python
#Stacking (vstack, hstack, dstack)
import numpy as np

a = np.array([1,2])
b = np.array([3,4])
print(np.vstack((a,b)))
print(np.hstack((a,b)))

c = np.array([[1,2],[3,4]])
d = np.array(([[5,6],[7,8]]))
result = np.dstack((c,d))
print("&&&&&&&&&&&&&&&&&")
print(result)
```

```python
#Splitting (split, hsplit, vsplit)
array = np.array([[1,2,3],[4,5,6]])
print(np.hsplit(array,3)) #split into 3 columns
array = np.array([[1,23],[4,56]])
print(np.vsplit(array,1))
```

```python
#Vectorization
arr = np.array([1,2,3])
print(arr * 2)
```

```python
#slicing
arr = np.array([10,20,30,40,50])
print(arr[1:4])
print(arr[:3])
print(arr[3:])
print(arr[::])
#indexing
arr1 = np.array([[1,2],[3,4]])
print(arr1[1,0])
```

```python
#masking
arr = np.array([1,2,3,4,5])
mask = arr >3
print(arr[mask])
```

```python
#sorting
arr = np.array([5,2,9,1])
print(np.sort(arr))
```

# **Pandas functions:**
Data Cleaning, Data types Handling, Data Exploration, Concating, merging, sorting, Aggregation, Group by, Pivot_table, Binning.


```python
import pandas as pd

df = pd.DataFrame({'Name':['Alice','Bob',None], 'Age':[25,None,30]})

# Drop rows with missing values
df_cleaned = df.dropna()
print(df_cleaned)

# Fill missing values
df_filled = df.fillna({'Name':'Unknow', 'Age':df['Age'].mean()})
print(df_filled)

```

```python
#Data Types Handling
df['Age'] = df['Age'].astype('float')
print(df.dtypes)
```

```python
#Data Exploration
print(df.head())
print(df.describe())
print(df.info())
```

```python
#Concatenating
df1 = pd.DataFrame({'A':[1,2]})
df2 = pd.DataFrame({'A':[3,4]})
df_concat = pd.concat([df1,df2], ignore_index=True)
print(df_concat)
```

```python
#Merging
left = pd.DataFrame({'ID':[1,2], 'Name':['Alice','Bob']})
right = pd.DataFrame({'ID':[1,2], 'Score':[90,85]})
df_merged = pd.merge(left,right, on='ID')
print(df_merged)
```

```python
#Sorting
df_sorted = df.sort_values(by='Age', ascending=True)
print(df_sorted)
```

```python
#Aggregation

df = pd.DataFrame({'Team': ['A', 'A', 'B'], 'Score': [10, 20, 30]})
total_score = df['Score'].sum()
print(total_score)
mean_score = df['Score'].mean()
print(mean_score)
```

```python
#Group By
grouped = df.groupby('Team')['Score'].sum()
print(grouped)
```

```python
#Pivot Table


df = pd.DataFrame({'Team': ['A', 'A', 'B', 'B'],'Player': ['P1', 'P2', 'P3', 'P4'], 'Score': [10, 15, 20, 25]})
pivot = df.pivot_table(values='Score', index='Team', columns='Player', fill_value=0)
print(pivot)
```

```python
#Binning

ages = pd.Series([15, 25, 35, 45, 55])
bins = [0, 20, 40, 60]
labels = ['Young', 'Adult', 'Senior']

age_groups = pd.cut(ages, bins=bins, labels=labels)
print(age_groups)
```

# **Data Visualisation:**
Different kinds of plots - Univariate(Cat, Cont), Bivariate (Cat-Cat, Cat-Cont, Cont-Cont), Multivariate.

# **ðŸ”¹ Univariate Analysis**
Univariate analysis focuses on a single variable:

Categorical: A count plot shows the frequency of each category.
Continuous: A histogram (with KDE) shows the distribution of a numeric variable.

# **ðŸ”¹ Bivariate Analysis**
Bivariate analysis explores relationships between two variables:

Categorical vs Categorical: A grouped bar chart (count plot with hue) shows how categories relate.
Categorical vs Continuous: A box plot shows the distribution of a continuous variable across categories.
Continuous vs Continuous: A scatter plot shows the relationship between two numeric variables.

# **ðŸ”¹ Multivariate Analysis**
Multivariate analysis involves more than two variables:

A scatter plot with color (hue) and marker style (style) encodes additional categorical variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Score': np.random.normal(loc=70, scale=10, size=100),
    'Age': np.random.randint(18, 60, size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100)
})

# Set the style for seaborn
sns.set(style="whitegrid")

# Univariate Analysis
plt.figure(figsize=(12, 6))

# Categorical
plt.subplot(1, 2, 1)
sns.countplot(x='Category', data=data)
plt.title('Univariate - Categorical')

# Continuous
plt.subplot(1, 2, 2)
sns.histplot(data['Score'], kde=True)
plt.title('Univariate - Continuous')

plt.tight_layout()
plt.show()

# Bivariate Analysis
plt.figure(figsize=(18, 6))

# Categorical-Categorical
plt.subplot(1, 3, 1)
sns.countplot(x='Category', hue='Gender', data=data)
plt.title('Bivariate - Categorical vs Categorical')

# Categorical-Continuous
plt.subplot(1, 3, 2)
sns.boxplot(x='Category', y='Score', data=data)
plt.title('Bivariate - Categorical vs Continuous')

# Continuous-Continuous
plt.subplot(1, 3, 3)
sns.scatterplot(x='Age', y='Score', hue='Gender', data=data)
plt.title('Bivariate - Continuous vs Continuous')

plt.tight_layout()
plt.show()

# Multivariate Analysis
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Score', hue='Gender', style='Category', data=data)
plt.title('Multivariate Analysis')
plt.tight_layout()
plt.show()
```
