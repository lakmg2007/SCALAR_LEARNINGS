# **The Central Limit Theorem states that:**

The sampling distribution of the sample mean (or sum) of a large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution, regardless of the original distribution of the data.

Key Concepts
Sampling Distribution:

When you take multiple samples from a population and compute their means, those means form a new distribution  the sampling distribution.
Original Distribution Doesn't Matter:

The population could be skewed, uniform, exponential, etc. But the distribution of sample means will tend toward a normal distribution as the sample size increases.

# **Example**
Imagine you're measuring the time it takes to load a webpage. The times are skewed (some pages take much longer). If you take the average load time from many samples of 30 pages each, the distribution of those averages will look normal  thanks to CLT.

# **Left Plot: Skewed Population Distribution**
The data is drawn from an exponential distribution, which is highly skewed.
This represents a real-world scenario where data isn't normally distributed (e.g., time between website visits).
## **Right Plot: Distribution of Sample Means**
We took 1,000 samples, each of size 30, and computed their means.
Despite the original data being skewed, the distribution of sample means is approximately normal.
This is the Central Limit Theorem in action!

image.png

# **Interpretation**
# ** Sample Size = 5**
The distribution of sample means is still skewed.
Small samples retain characteristics of the original (exponential) distribution.
# ** Sample Size = 30**
The distribution begins to smooth out and resemble a normal distribution.
This is the typical threshold where CLT starts to take effect.
# ** Sample Size = 100**
The distribution of sample means is very close to normal.
Larger samples average out the variability and skewness.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Generate a skewed population distribution
population = np.random.exponential(scale=2.0, size=10000)

# Parameters
sample_size = 30
num_samples = 1000
sample_means = []

# Draw samples and compute means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot the original population distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='skyblue')
plt.title('Skewed Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=50, kde=True, color='salmon')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
