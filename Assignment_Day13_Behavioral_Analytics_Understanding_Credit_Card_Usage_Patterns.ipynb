# **Question 1:** According to a survey conducted on car owners, it was determined that

# 60% of owners have only one car,
# 28% have two cars, and
# 12% have three or more cars.

# Suppose Ram conducted his own survey within his residential society, and found that

# 73 owners have only one car,
# 38 owners have two cars, and
# 18 owners have three or more cars.
# Determine whether Ram's survey supports the original one, with a significance level of 0.05.

# **Why Use the Chi-Square Goodness-of-Fit Test?**
# **Type of Data:**

# The data is categorical: car ownership is divided into discrete categories (1 car, 2 cars, 3+ cars).
# We're counting how many individuals fall into each category.
Goal of the Test:

# To determine whether the observed frequencies match the expected frequencies.
# It helps test the null hypothesis: that Ram's data follows the same distribution as the original survey.
# No Assumptions About Distribution Shape:

# Unlike tests that assume normality (like t-tests), the chi-square test doesn't require the data to be normally distributed.
# Simple and Effective:

# It’s a widely used and straightforward method for comparing distributions in categorical data.


# In a chi-square test, we compare actual observed counts (from Ram's survey) to expected counts (based on the original survey proportions). But the original survey gives us percentages, not raw numbers.

# To make a valid comparison, we need to scale those percentages to match the total number of people in Ram's survey.

# Ram surveyed 129 people (73 + 38 + 18).
# Original survey says:
# 60% should have one car → 0.60 × 129 = 77.4
# 28% should have two cars → 0.28×129=36.12
# 12% should have three or more cars → 0.12×129=15.48

```python
from scipy.stats import chisquare # Statistical test (chistat, pvalue)
from scipy.stats import chi2

observed = [73, 38,18]
expected_proportions = [0.60, 0.28, 0.12]

# Total number of surveyed owners
total = sum(observed)


# Calculate expected frequencies based on proportions
expected = [p * total for p in expected_proportions]


chi_stat, p_value = chisquare(observed,expected)

print("p_value:",p_value)
print("chi_stat:",chi_stat)


# Conclusion
if p_value < 0.05:
    print("Ram's survey does NOT support the original distribution (statistically significant difference).")
else:
    print("Ram's survey supports the original distribution (no statistically significant difference).")
```

# **Question 2 :** A national survey was conducted to obtain information on the alcohol consumption patterns of U.S. adults by marital status.
# A random sample of 1772 residents, aged 18 and older, yielded the data displayed in Table below:

image.png

Test whether Marital status and alcohol consumption are associated with a 5% significance level.

Choose the correct option below :

# **why Chi-Square Test of Independence considered?**
# **Purpose of the Test**
# The test is used to determine whether there is a statistically significant association between two categorical variables. In your case:

# Variable 1: Marital Status (Single, Married, Widowed, Divorced)
# Variable 2: Alcohol Consumption (Abstain, 1–60 drinks/month, Over 60 drinks/month)

# **Why It’s Suitable**
#**1. Both variables are categorical:**

# Marital status and alcohol consumption are not numerical but fall into distinct categories.
#**2. We’re testing for association: **bold text** **

# The goal is to see if alcohol consumption patterns differ depending on marital status.
#**3. Data is in a contingency table: **bold text** **

#The observed frequencies are organized in a matrix format (rows and columns), which is exactly what the chi-square test of independence is designed for.
# **4. No assumptions about distribution:**

#Unlike parametric tests, this test doesn’t assume a normal distribution of the data.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed frequency table
# Rows: Single, Married, Widowed, Divorced
# Columns: Abstain, 1–60 drinks/month, Over 60 drinks/month
observed = np.array([
    [67, 213, 74],    # Single
    [411, 633, 129],  # Married
    [85, 51, 7],      # Widowed
    [27, 60, 15]      # Divorced
])

# Perform chi-square test of independence
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

# Print results
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p_value:.4f}")

# Conclusion at 5% significance level
if p_value < 0.05:
    print("Reject the null hypothesis: Marital status and alcohol consumption are associated.")
else:
    print("Fail to reject the null hypothesis: No significant association between marital status and alcohol consumption.")
```

# **Question : 6** The United States is divided into four regions: Northeast, Midwest, South and West regions.

# Independent random samples of households in these regions yielded the data on last year’s energy consumptions.

# a = [13, 8, 11, 12, 11]
# b = [15, 10, 16, 11, 13, 10]
# c = [5, 11, 9, 5]
# d = [8, 10, 6, 5, 7]

image.png
# At a 5% significance level, does the data provide sufficient evidence to conclude that there is a difference in last year’s mean energy consumption
# by households among the four regions?

# Choose the correct option below :

# **why one-way ANOVA test  is considered?**

# A one-way ANOVA (Analysis of Variance) test is used in this context because you're comparing the means of energy consumption across more than two independent groups—specifically, the four U.S. regions:

#Northeast
#Midwest
#South
#West

#Here’s why one-way ANOVA is appropriate:
# **Purpose of One-Way ANOVA**
# To test whether there are statistically significant differences between the means of three or more independent groups.

# **Why Not Use t-tests Instead?**
# You could use t-tests to compare two groups at a time.
# But with four groups, you'd need multiple t-tests, which increases the risk of Type I error (false positives).

# ANOVA controls for this by testing all groups simultaneously.

#**Assumptions of One-Way ANOVA**
#1. Independence: Samples are independent.
#2. Normality: Data in each group is approximately normally distributed.
#3. Homogeneity of variances: Variances across groups are roughly equal.

#Interpretation
#Null Hypothesis (H₀): All group means are equal.
#Alternative Hypothesis (H₁): At least one group mean is different.

#If the p-value < significance level (e.g., 0.05), you reject H₀ and conclude that at least one region differs in mean energy consumption.





```python
from scipy.stats import f_oneway

# Sample data for each region
northeast = [13, 8, 11, 12, 11]
midwest = [15, 10, 16, 11, 13, 10]
south = [5, 11, 9, 5]
west = [8, 10, 6, 5, 7]

# Perform one-way ANOVA test
f_statistic, p_value = f_oneway(northeast, midwest, south, west)

# Print results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion at 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in mean energy consumption among the four regions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in mean energy consumption among the four regions.")
```

# Those variables, along with their categories are:

        Variable A: PoliticalOpinions

            Disagree,
            Strongly Disagree
        Variable B: DemographicInfo (Age Group)
            18-25,
            26-35,
            36-50
# The goal is to determine if there is a significant association between the opinions on the political issue and demographic characteristics, specifically age groups.

# In this scenario, what is the degrees of freedom for the chi-square test of independence?

# The Chi-square test of independence is used in your scenario because you're analyzing the relationship between two categorical variables:

# Variable A: PoliticalOpinions (Strongly Agree, Agree, Disagree, Strongly Disagree)
# Variable B: Age Group (18–25, 26–35, 36–50)

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table
# Rows: PoliticalOpinions (Strongly Agree, Agree, Disagree, Strongly Disagree)
# Columns: Age Groups (18–25, 26–35, 36–50)
# Replace these values with your actual survey data
observed = np.array([
    [2, 30, 25],  # Strongly Agree
    [25, 3, 30],  # Agree
    [15, 20, 25],  # Disagree
    [10, 1, 20]   # Strongly Disagree
])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Print results
print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

# Conclusion at 5% significance level
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between political opinions and age group.")
else:
    print("Fail to reject the null hypothesis: No significant association between political opinions and age group.")
```
