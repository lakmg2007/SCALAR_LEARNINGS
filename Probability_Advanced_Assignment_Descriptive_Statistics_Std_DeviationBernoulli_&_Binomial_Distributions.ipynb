```python
#Question1 : In a factory, the probability of producing a defective bulb is 0.25.
# A sample of 40 bulbs is collected.What is the probability that exactly 10 bulbs are defective?

#Each bulb has two outcomes: defective or not defective.
# The probability of a defective bulb is constant: 𝑝=0.25
# The sample size is fixed: n=40
# We want the probability of exactly k=10 defective bulbs.

from scipy.stats import binom
n=40
k=10
p=0.25
probability = binom.pmf(k, n, p)
print(f"probability of exactly k=10 defective bulbs: {probability:.4f}")
```

#Question 2: A packaging plant fills bags with cement.The weight of the cement bags can be modelled by a normal distribution with a mean of 50kg and a standard deviation of 2 kg.Find the weight x that is exceeded by 99% of the bags.

```python
mean = 50
std = 2
percentile = 0.01  # Since we want the weight that 99% exceed => 1%
from scipy.stats import norm

ppf_99 = norm.ppf(percentile, loc=mean, scale=std)
ppf_99
```

# Question 5: The mean weight of 4 students is 40 kg. If the weight of the teacher is included, the average becomes 42 kg. What will be the weight of the teacher?

```python
mean_students = 40      # mean weight of 4 students
num_students = 4
mean_with_teacher = 42  # mean weight with teacher included
total_people = 5        # 4 students + 1 teacher

# Step 1: Total weight of students
total_students_weight = mean_students * num_students

# Step 2: Total weight including teacher
total_group_weight = mean_with_teacher * total_people

# Step 3: Teacher's weight
teacher_weight = total_group_weight - total_students_weight

print(f"The teacher's weight is: {teacher_weight} kg")
```

#Question 6: The salary of employees (in thousands) is given as: 40, 42, 45, 50, x. If the mean salary is 45. What will be the median of the salaries?

image.png

image.png

```python
import statistics

# Step 1: Known salaries and mean
salaries = [40, 42, 45, 50]  # in thousands
mean_salary = 45

# Step 2: Calculate the missing salary (x)
x = mean_salary * 5 - sum(salaries)
salaries.append(x)

# Step 3: Sort and find the median
salaries.sort()
median_salary = statistics.median(salaries)

# Output
print(f"Calculated missing salary (x): {x}k")
print(f"Sorted salaries: {salaries}")
print(f"Median salary: {median_salary}k")
```
